# creating a clustering model

```python
# to convert string into numeric values 
X=pd.get_dummies(df,dtype='int64')
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_X=scaler.fit_transform(X)
```

# model

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=2)
```

```python
pred=model.fit_predict(scaled_X) # similar to fit_transform
```

```python
model=KMeans(n_clusters=2,n_init='auto') #to overcome warnings
```

```python
pred=model.fit_predict(scaled_X)
```

```python
pred
```

* here, prediction value like 0 and 1 keep changing as we run 'fit_predict' statement and these predicted values has no meaning (to avoid this we can set a 'random_state' parameter)

```python
X['Cluster']=pred #adding a predicted value to dataset
```

```python
X
```

```python
X.corr()
```

```python
X.corr()['Cluster'] # to see how every features related to predicted output
```

```python
X.corr()['Cluster'].sort_values() #to see most important features
```

```python
plt.figure(figsize=(12,6),dpi=200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
plt.xticks(rotation=90);
```

# measuring a best 'n_cluster' value

* 1.measure the sum of squares distances(SSD) from points to cluster centers

```python
ssd=[]

for n in range(1,10):
    model=KMeans(n_clusters=n,n_init='auto')
    model.fit(scaled_X)
    
    ssd.append(model.inertia_) # calculates ssd from cluster center to data points
```

```python
ssd
```

```python
plt.plot(range(1,10),ssd,'o--')
```

# there is no much drop of 'SSD' after k=6 so it would be a ideal value

```python
pd.Series(ssd)
```
