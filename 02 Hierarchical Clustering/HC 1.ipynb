```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
df=pd.read_csv("cluster_mpg.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies=pd.get_dummies(df.drop('name',axis=1))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler() #scales data between 0 and 1
```

```python
scale_data=scaler.fit_transform(df_w_dummies)
```

```python
scale_data
```

```python
scale_df=pd.DataFrame(data=scale_data,columns=df_w_dummies.columns)
```

```python
scale_df
```

```python
sns.heatmap(scale_df)
```

```python
sns.clustermap(data=scale_df)
```

```python
sns.clustermap(data=scale_df,col_cluster=False)
```

# main split is 'origin_usa' and 2nd is 'cylinders'

```python
model=AgglomerativeClustering(n_clusters=4)
```

```python
labels=model.fit_predict(scale_df)
```

```python
labels
```

```python
sns.scatterplot(data=scale_df,x='mpg',y='weight',hue=labels,palette='viridis')
```

```python
len(scale_df.columns)
```

```python
np.sqrt(10) #MAX distance possible between any two points (only when we use 'Min_Max_scaling')
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=0)
```

```python
labels=model.fit_predict(scale_df)
```

```python
labels
```

```python
len(labels)
```

# no merging of any points because 'distance_threshold =0' unique cluster for every point

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix=hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

Linkage matrix shows:-
* 1 & 2 col shows clusters joining together(labels) ,3 col shows distance between two clusters ,4 col shows number of points in that cluster ,as we go down every point will lie on single cluster 

```python
plt.figure(figsize=(20,10))
dendro=dendrogram(linkage_matrix)
```

* this above visualization may not be useful so we can cut it out at any level using parameter 'truncate_mode'

```python
plt.figure(figsize=(20,10))
dendro=dendrogram(linkage_matrix,truncate_mode='lastp',p=10)
```

```python
plt.figure(figsize=(20,10))
dendro=dendrogram(linkage_matrix,truncate_mode='level',p=3)
```
